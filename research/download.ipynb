```python
import os
import requests
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
i = 0
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
URL = 'https://api.isic-archive.com/api/v2'
```

we define dictionary flatting function

```python
def flatten_dict(d, sep='.'):
    [flat_dict] = pd.json_normalize(d, sep=sep).to_dict(orient='records')
    return flat_dict
```

we create a table

```python
data = pd.DataFrame()
```

we make a requests

```python
dataportion = requests.get(f'{URL}/images?limit=1000&offset=0').json()
```

```python
while dataportion and i < 10:
    results = dataportion['results']
    
    for r in results:
        data = data.append(flatten_dict(r), ignore_index=True)
        
    
    dataportion = requests.get(dataportion['next']).json() if dataportion['next'] else None
    i += 1
```
